# Task
Apply SHAP to analyze overall feature contributions in a tree-based model trained on the Iris dataset and generate summary plots.

## Load and prepare data

### Subtask:
Load the Iris dataset and prepare it for training a model.


**Reasoning**:
The subtask requires loading the Iris dataset and creating a Pandas DataFrame. This can be achieved by importing `load_iris`, loading the data, and then creating the DataFrame using the feature names.



```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target

df_iris = pd.DataFrame(X, columns=iris.feature_names)

 
```

## Train a tree-based model

### Subtask:
Train a tree-based model on the Iris dataset.


**Reasoning**:
Train a RandomForestClassifier model on the Iris dataset.



```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X, y)
 
```

## Compute shap values

### Subtask:
Compute SHAP values for the trained model.


**Reasoning**:
Compute SHAP values for the trained model.



```python
import shap

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)
 
```

## Generate shap summary plots

### Subtask:
Generate SHAP summary plots to visualize overall feature contributions.


**Reasoning**:
Generate a SHAP summary plot using the computed SHAP values and the feature data and display the plot.



```python
shap.summary_plot(shap_values, X, feature_names=iris.feature_names)
 
```

## Summary:
